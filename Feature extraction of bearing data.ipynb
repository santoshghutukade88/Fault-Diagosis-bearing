```python
import numpy as np
import pandas as pd
import os
import math
import stat
```

```python
allfolders=os.listdir('E:\PYTHON\projects\Data set')
```

```python
path='E:\PYTHON\projects\Data set'
```

```python
allfiles=os.listdir('E:\PYTHON\projects\Data set\Clearance')
```

```python
final1=pd.DataFrame(columns=('mean','kurtosis','skewness','variance','STD','fault'))
```

```python
z1=[]
z2=[]
z3=[]
z4=[]
z5=[]
z6=[]
for i in allfolders: 
    allfiles=os.listdir('E:\PYTHON\projects\Data set'+'\\'+ i)
    for j in allfiles:
        df=pd.read_excel(path + '\\'+i+'\\'+ j)
        z1.append(df.iloc[:,1].mean())
        z2.append(df.iloc[:,1].kurtosis())
        z3.append(df.iloc[:,1].skew())
        z4.append(df.iloc[:,1].var())
        z5.append(df.iloc[:,1].std())
        z6.append(i)
        
final=pd.DataFrame({'mean':z1,'kurtosis':z2,'skewness':z3,'variance':z4,'STD':z5,'fault':z6})
```
